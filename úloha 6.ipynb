```python
# Načtení knihoven a deklarování proměnných 
from turtle import position
import pyautogui
import time
import numpy as np
import random as r
pozice = []
seminko = 0
```

```python
print('Pohybujte myší po obrazovce pro vygenerování semínka.')
for i in range(20):
    x, y = pyautogui.position()
    pozice.append((x, y))
    print(f'Pozice myši '+str(x) + ' '+str(y) + ' zaznamenána.')
    for i in pozice:
        seminko += np.math.factorial(sum(i))
    time.sleep(0.2)
print(f'Náhodně vygenerované semínko: {seminko}')
```
