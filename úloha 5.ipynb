```python
# Načtení knihoven a deklarování tří funkcí
import numpy as np
import matplotlib.pyplot as plt
import timeit

fce1 = lambda x: 3*x**3 - 5*x**2 + 5*x - 8
fce2 = lambda x: x + np.exp(2*x) - 5
fce3 = lambda x: np.cos((x+9))
```

```python
def bisekce(f, a0, b0, eps=1E-10):
    a, b = a0, b0
    while b - a > 2*eps:
        x = (a + b)/2
        if f(a)*f(x) < 0:
            b = x
        else:
            a = x
    return x

def newton(f, a0, b0, delta=1E-10):
    a, b = a0, b0
    def df(x, h = 1E-10):
        return (f(x+h)-f(x-h))/(2*h)
    xnew = (a+b)/2
    xold = a
    while abs(xnew - xold) > delta:
        xold = xnew
        xnew = xold - f(xold)/df(xold)
    return xnew

def graf(f, a0, b0, x, title):
    print("root: ", x)
    plt.title(title)
    plt.plot(x, f(x), "bo")                                              
    plt.plot(np.linspace(a0, b0, 100), f(np.linspace(a0, b0, 100)), "r")
    plt.axhline(y=0)
    plt.show()
```

```python
# Funkce 1
a, b = 0 , 4
graf(fce1, a, b, bisekce(fce1, a, b), "Bisekce: 3*x**3 - 5*x**2 + 5*x - 8")
graf(fce1, a, b, newton(fce1, a, b), "Newton: 3*x**3 - 5*x**2 + 5*x - 8")
time1 = timeit.timeit(lambda: bisekce(fce2, a, b), number=5000)
time2 = timeit.timeit(lambda: newton(fce2, a, b), number=5000)
print(f"BISECTION: {time1/5000}s, NEWTON: {time2/5000}s")
```

```python
# Funkce 2
a, b = 0 , 2
graf(fce1, a, b, bisekce(fce1, a, b), "Bisekce: x + e^(2*x) - 5")
graf(fce1, a, b, newton(fce1, a, b), "Newton: x + e^(2*x) - 5")
time1 = timeit.timeit(lambda: bisekce(fce2, a, b), number=5000)
time2 = timeit.timeit(lambda: newton(fce2, a, b), number=5000)
print(f"BISECTION: {time1/5000}s, NEWTON: {time2/5000}s")
```
